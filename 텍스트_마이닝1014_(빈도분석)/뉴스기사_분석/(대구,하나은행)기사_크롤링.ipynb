```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import random


# 데이터를 담을 리스트와 집합 초기화
titles = set()  # 중복 방지를 위해 set 사용
descriptions = []

# 스크롤을 계속 내려서 데이터를 수집하는 코드
while True:
    try:
        # 페이지의 끝까지 스크롤을 내림
        body = driver.find_element(By.CSS_SELECTOR, 'body')
        body.send_keys(Keys.PAGE_DOWN)  # 페이지 아래로 스크롤
        time.sleep(random.uniform(1, 2))  # 1~2초 사이의 랜덤 대기 시간

        # 페이지 소스를 가져와 BeautifulSoup으로 파싱
        html = BeautifulSoup(driver.page_source, 'html.parser')

        # 뉴스 리스트를 가져옴
        news_list = html.find('ul', class_='list_news _infinite_list')

        # news_list가 None이 아닐 때만 진행
        if news_list is not None:
            # 각 뉴스의 li 요소들을 가져옴
            li_elements = news_list.find_all('li', class_='bx')

            for li in li_elements:
                # 뉴스 제목과 본문을 한 번에 추출
                title_tag = li.find('a', class_='news_tit')
                description_tag = li.find('a', class_='api_txt_lines dsc_txt_wrap')

                # 제목과 본문이 모두 있을 때만 처리
                if title_tag and description_tag:
                    title = title_tag.get_text(strip=True)
                    description = description_tag.get_text(strip=True)

                    # 중복된 제목이 아닐 경우 추가
                    if title not in titles:
                        titles.add(title)  # 제목을 set에 추가
                        descriptions.append(description)  # 본문 리스트에 추가

        # 특정 종료 조건 (예: 100개 이상의 제목 수집 시 종료)
        if len(titles) >= 100:  # 100개 이상 수집 시 종료
            break

    except Exception as e:
        print(f"에러 발생: {e}")
        break

# pandas 데이터프레임으로 변환
df = pd.DataFrame({
    '제목': list(titles),       # 제목 리스트
    '본문': descriptions  # 본문 리스트
})

# 데이터프레임 출력
df


```

```python
driver.get('https://search.naver.com/search.naver?sm=tab_hty.top&where=news&ssc=tab.news.all&query=%ED%95%98%EB%82%98%EC%9D%80%ED%96%89&oquery=im%EB%B1%85%ED%81%AC&tqi=iwypBwqo1SCssuE%2FJOVssssstf8-345257&nso=so%3Add%2Cp%3Aall&mynews=0&office_category=0&office_section_code=0&office_type=0&pd=0&photo=0&service_area=0&sort=1')
```

```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import random


# 데이터를 담을 리스트와 집합 초기화
titles = set()  # 중복 방지를 위해 set 사용
descriptions = []

# 스크롤을 계속 내려서 데이터를 수집하는 코드
while True:
    try:
        # 페이지의 끝까지 스크롤을 내림
        body = driver.find_element(By.CSS_SELECTOR, 'body')
        body.send_keys(Keys.PAGE_DOWN)  # 페이지 아래로 스크롤
        time.sleep(random.uniform(1, 2))  # 1~2초 사이의 랜덤 대기 시간

        # 페이지 소스를 가져와 BeautifulSoup으로 파싱
        html = BeautifulSoup(driver.page_source, 'html.parser')

        # 뉴스 리스트를 가져옴
        news_list = html.find('ul', class_='list_news _infinite_list')

        # news_list가 None이 아닐 때만 진행
        if news_list is not None:
            # 각 뉴스의 li 요소들을 가져옴
            li_elements = news_list.find_all('li', class_='bx')

            for li in li_elements:
                # 뉴스 제목과 본문을 한 번에 추출
                title_tag = li.find('a', class_='news_tit')
                description_tag = li.find('a', class_='api_txt_lines dsc_txt_wrap')

                # 제목과 본문이 모두 있을 때만 처리
                if title_tag and description_tag:
                    title = title_tag.get_text(strip=True)
                    description = description_tag.get_text(strip=True)

                    # 중복된 제목이 아닐 경우 추가
                    if title not in titles:
                        titles.add(title)  # 제목을 set에 추가
                        descriptions.append(description)  # 본문 리스트에 추가

        # 특정 종료 조건 (예: 100개 이상의 제목 수집 시 종료)
        if len(titles) >= 100:  # 100개 이상 수집 시 종료
            break

    except Exception as e:
        print(f"에러 발생: {e}")
        break

# pandas 데이터프레임으로 변환
df = pd.DataFrame({
    '제목': list(titles),       # 제목 리스트
    '본문': descriptions  # 본문 리스트
})

# 데이터프레임 출력
df

```

```python
# 중복된 행 확인
duplicates = df.duplicated()

# 중복된 행의 수
num_duplicates = duplicates.sum()

# 중복된 행 출력
duplicate_rows = df[duplicates]

print(f"중복된 행의 수: {num_duplicates}")
print("중복된 행:")
print(duplicate_rows)
```
