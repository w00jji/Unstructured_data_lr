```python
# 기존의 Okt 및 Counter 라이브러리 임포트는 생략합니다.

# 이명박 데이터 처리
lee_df = pd.DataFrame(lee_counter.most_common(), columns=['단어', '빈도'])
lee_df['대통령'] = '이명박'

# 노무현 데이터 처리
no_df = pd.DataFrame(no_counter.most_common(), columns=['단어', '빈도'])
no_df['대통령'] = '노무현'

# 문재인, 박근혜 데이터 처리 (기존 코드와 동일)
moon_df = pd.DataFrame(moon_counter.most_common(), columns=['단어', '빈도'])
moon_df['대통령'] = '문재인'

park_df = pd.DataFrame(park_counter.most_common(), columns=['단어', '빈도'])
park_df['대통령'] = '박근혜'

# 모든 데이터를 하나의 데이터프레임으로 병합
df = pd.concat([lee_df, no_df, moon_df, park_df], ignore_index=True)

# 피벗 테이블로 각 대통령별 빈도 비교
pivot_df = df.pivot_table(index='단어', columns='대통령', values='빈도').fillna(0)

# 각 대통령별 빈도 비율 계산 (스무딩 적용, 1을 더해서 나눔)
pivot_df['이명박 Ratio'] = (pivot_df['이명박'] + 1) / sum(pivot_df['이명박'] + 1)
pivot_df['노무현 Ratio'] = (pivot_df['노무현'] + 1) / sum(pivot_df['노무현'] + 1)
pivot_df['문재인 Ratio'] = (pivot_df['문재인'] + 1) / sum(pivot_df['문재인'] + 1)
pivot_df['박근혜 Ratio'] = (pivot_df['박근혜'] + 1) / sum(pivot_df['박근혜'] + 1)

# Odds Ratio 계산: 이명박이 다른 대통령들에 비해 상대적으로 많이 나온 단어
pivot_df['이명박 vs 노무현'] = pivot_df['이명박 Ratio'] / pivot_df['노무현 Ratio']
pivot_df['이명박 vs 문재인'] = pivot_df['이명박 Ratio'] / pivot_df['문재인 Ratio']
pivot_df['이명박 vs 박근혜'] = pivot_df['이명박 Ratio'] / pivot_df['박근혜 Ratio']

# 이명박이 상대적으로 많이 언급된 단어 10개 정렬
pivot_df['이명박 Overall'] = pivot_df[['이명박 vs 노무현', '이명박 vs 문재인', '이명박 vs 박근혜']].mean(axis=1)
pivot_df = pivot_df.sort_values('이명박 Overall', ascending=False)

# 이명박 대통령의 빈도가 다른 대통령보다 높은 단어 10개 출력
lee_top10_vs_others = pivot_df.head(10)
lee_top10_vs_others
```

## 노무현 대통령이 다른 대통령에 비해 많이 나온 단어 10개를 찾아보세요

```python
total = []
df = pd.read_csv("/content/drive/MyDrive/IMbank_텍스트마이닝/텍스트마이닝_1014/speeches_presidents.csv")

for i in range(len(df)):
    a = df.iloc[i]
    president = a['president']
    text = okt.nouns(a['value'])
    box = []
    for j in text:
        if len(j) >= 2:
            box.append(j)
    counter = Counter(box).most_common()
    sample = pd.DataFrame(counter, columns = ['단어', '빈도'])
    if president == "이명박":
        sample['대통령'] = "이명박"
    else:
        sample['대통령'] = "그 외"
    total.append(sample)
```

```python
df = pd.concat(total, ignore_index=True)
df2 = df.pivot_table(index = "단어", columns = "대통령",
                     values = "빈도").fillna(0)

df2["그 외 Ratio"] = (df2['그 외'] + 1) / sum(df2['그 외'] + 1)
df2["이명박 Ratio"] = (df2['이명박'] + 1) / sum(df2['이명박'] + 1)
df2['Odds Ratio'] = df2["이명박 Ratio"] / df2["그 외 Ratio"]
df2.sort_values("Odds Ratio", ascending = False)
```

```python
df2['Log Odds Ratio'] = np.log(df2['Odds Ratio'])
df2 = df2.sort_values('Log Odds Ratio', ascending = False)

a = df2.head(10)
b = df2.tail(10)

df3 = pd.concat([a, b])

plt.barh(a.index, a['Log Odds Ratio'], color = 'red', label = '이명박')
plt.barh(b.index, b['Log Odds Ratio'], color = 'blue', label = '그 외')
plt.grid(axis = 'x')
plt.legend()
plt.show()
```

```python
okt = Okt()
def make_nouns(x):
    nouns = okt.nouns(x) # 명사 추출 후 저장
    box = []
    for i in nouns:
        if len(i) >= 2: # 길이가 2 이상만 저장
            box.append(i)
    return box

df = pd.read_csv("/content/drive/MyDrive/IMbank_텍스트마이닝/텍스트마이닝_1014/speeches_presidents.csv")
df['value_nouns'] = df['value'].apply(make_nouns)
```

```python
df
```

```python
unique_words = set()
for i in df['value_nouns']:
    #print(i) # 총 4명의 대통령 연설문의 명사를 가져옴
    for j in i:
        #print(j) # 명사를 하나씩 가져옴
        unique_words.add(j) # 중복되지 않은 단어 저장

print(unique_words)
```

```python
# DF : 단어가 몇개의 문서에 출현했는지
DF_dic = {}
for i in unique_words:
    # print(i) # 고유한 단어 중복되지 않은 단어들만 추출
    count = 0
    for j in df['value_nouns']:
        if i in j: # 만약 i가 j라는 연설문에 단어가 있다면 딕셔너리로 count +1 증가하면서 저장, 몇번 출현했는지 계산
            count += 1
    DF_dic[i] = count

DF_dic
```

```python
df2=df.explode('value_nouns')[['president', 'value_nouns']]
df3 =df2.groupby(['president', 'value_nouns']).size().reset_index(name='TF') # 빈도
df3
```

```python
def make_DF(x):
    return DF_dic[x]

df3['DF'] = df3['value_nouns'].apply(make_DF) # 몇명의 대통령이 사용했는지의 수
df3
```

```python
# IDF: 총 문서  - 4/DF
df3['IDF'] = 4 / df3['DF']
df3['TF-IDF'] = df3['TF'] * df3['IDF'] # 어떤 단어가 가장 많이 사용됐는지 수치 Ex) 자소서 항목에서도 많이 사용됨
df3
```
