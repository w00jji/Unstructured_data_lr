```python
import time
from bs4 import BeautifulSoup
import pandas as pd

start_page = 1
end_page = 10

data = []

for page in range(start_page, end_page + 1):
    url = f"https://land.naver.com/news/headline.naver?page={page}"
    driver.get(url)
    time.sleep(2)  

    # 페이지 소스에서 뉴스 아이템 리스트 추출
    html = BeautifulSoup(driver.page_source, 'html.parser')
    news_items = html.find("ul", class_='land_news_list').find_all("li", class_="news_item")

    # 각 뉴스 아이템에서 제목과 설명 추출
    for item in news_items:
        title = item.find("p", class_="title").get_text(strip=True)
        description = item.find_all("p", class_="description")[0].get_text(strip=True)
        
        
        data.append({"제목": title, "본문": description})


df = pd.DataFrame(data)

df
```

```python
# 최근 기준 2024/10/15까지의 기사

# DataFrame을 탭 구분 텍스트 파일로 저장
df.to_csv('news_data(부동산).txt', sep='\t', index=False, encoding='utf-8-sig')

```
